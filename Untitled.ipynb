```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('housing.csv')

data.dropna(inplace = True)
#dropping zero rows keeping the row index the same 
data.info()


```

```python
from sklearn.model_selection import train_test_split
#splits the data into x containing all collumns but the median house data and y the med house data as an array
x = data.drop(['median_house_value'],axis = 1)
y = data['median_house_value']
```

```python
#this splits the training data from the test data using 20% of the arrays for the test data 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
#displaying a histogram and dropping the ocean proximity strings from the array for a correlation matrix.
train_data = x_train.join(y_train)
train_data.hist()
ocean_proximity = train_data.ocean_proximity
train_data = train_data.drop(['ocean_proximity'],axis = 1)

train_data.corr()
```

```python
#Heatmap with annotation of the correlation of variables.
sns.heatmap(train_data.corr(), annot =True, cmap = "YlGnBu")
```

```python
# taking the logaritm then histogram shows that for the negatively skewed data the logarithm of the variables has almost normal distrabution this would be a lognormal distrabution.
train_data['total_rooms'] = np.log(train_data['total_rooms'] )
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] )
train_data['population'] = np.log(train_data['population'] )
train_data['households'] = np.log(train_data['households'] )
train_data
train_data.hist(figsize = (15,8))
```

```python
#creates linear regression on the train data

from sklearn.linear_model import LinearRegression

x_train,y_train = train_data.drop(['median_house_value'],axis = 1),train_data['median_house_value']
reg = LinearRegression()
reg.fit(x_train,y_train)
score = reg.score(x_test,y_test)
print(score)


# Data scatter of predicted values
```

```python
#random forest regression model
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor()
forest.fit(x_train,y_train)
forest.score(x_test,y_test)
```
